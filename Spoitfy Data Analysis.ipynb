```python
# Top 10 songs with popularity greater than 90
popular = tracksdf.query('popularity > 90', inplace = False).sort_values('popularity', ascending =  False)
popular.head(10)
```

```python
#Changing the index 
tracksdf.set_index('release_date',inplace = True)
#Setting to datetime format
tracksdf.index = pd.to_datetime(tracksdf.index)
tracksdf.head(2)
```

```python
#Checking the artist at 20th position [Specific data from a specific row]
tracksdf[['artists']].iloc[19]
#index in pyhton start from 0
```

```python
#Converting time form milliseconds to seconds and dropping the orignal colummn
tracksdf['duration']=tracksdf['duration_ms'].apply(lambda x : round (x /1000))
tracksdf.drop('duration_ms', inplace = True, axis = 1)
tracksdf.head(2)
```

```python
# Creating first visualization using the relations in between columns
```

```python
df1 = tracksdf.drop(['key','mode'], axis = 1).corr(method = 'pearson')
plt.figure(figsize =(15,8))
heatmap = sns.heatmap(df1, annot = True, fmt = '.1g',vmin = -1, vmax=1, center = 0, cmap = 'inferno', linewidths  =1, linecolor = 'Black')
heatmap.set_title('Correaltion between variables using Heatmap')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)

plt.savefig('Heatmap of relations between columns.png', dpi = 1500)
```

```python
# TO display the corelations, we can now plot it in form of regression, here we use a part of dataset to display it
df2 = tracksdf.sample(int(0.004*len(tracksdf)))
print(len(df2))
```

```python
#Regression plot
plt.figure(figsize = (10,6))
sns.regplot(data = df2, y = 'loudness', x = 'energy', color = 'c' ).set(title = 'Loudness vs energy Correlation')
```

```python
#Regression plot
plt.figure(figsize = (10,6))
sns.regplot(data = df2, y = 'popularity', x = 'acousticness', color = 'g' ).set(title = 'Popularity vs Acousticness')
```

```python
#Extracting year from dates table
tracksdf['dates'] = tracksdf.index.get_level_values('release_date')
tracksdf.dates = pd.to_datetime(tracksdf.dates)
years = tracksdf.dates.dt.year
years
```

```python
plt.figure(figsize = (10,8))
sns.displot(years,aspect = 2, height = 5, kind='hist').set(title = 'Songs relased per year')

plt.savefig('Songs relased per year.png', dpi = 1000)
```

```python
#duration of songs over the years
duration = tracksdf['duration']
duration
```

```python
plt.figure(figsize = (60,20))
sns.barplot(x=years, y = duration)
plt.title('Songs duration over the years in seconds', fontsize = 50)
plt.xticks(rotation = 90)
plt.margins(0.01)

plt.savefig('Songs duration over the years in seconds.png', dpi = 1000)
```

```python
plt.figure(figsize = (30,10))
sns.lineplot(x=years, y = duration)
plt.title('Songs duration over the years in seconds', fontsize = 50)
plt.margins(0.01)
```

```python
plt.figure(figsize = (30,10))
sns.set_style(style='whitegrid')
sns.lineplot(x=years, y = duration)
plt.title('Songs duration over the years in seconds', fontsize = 50)
plt.margins(0.01)
```

```python
plt.figure(figsize = (30,10))
sns.set_style(style='whitegrid')
sns.lineplot(x=years, y = duration)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.title('Songs duration over the years in seconds', fontsize = 50)
plt.margins(0.01)
```

```python
plt.figure(figsize = (30,10))
sns.set_style(style='whitegrid')
sns.lineplot(x=years, y = duration)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.title('Songs duration over the years in seconds', fontsize = 50)
plt.margins(0.01)
plt.xlabel('Years', fontsize=60)
plt.ylabel('Duration', fontsize=60)


# saving the plot
plt.savefig('Songs duration over the years500.png', dpi = 1000)
```
